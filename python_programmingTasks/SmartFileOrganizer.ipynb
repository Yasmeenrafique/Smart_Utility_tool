## **Smart Utility Tool**

```python
# organize_files.py
import os
import shutil
from datetime import datetime

# ------------ USER SETTINGS ------------
# Change this to the folder you want to organize (use absolute path for safety)
FOLDER_PATH = r"D:\python_programmingTasks\Test files"

# Set DRY_RUN = True to see what will happen without moving files
DRY_RUN = False
# ---------------------------------------

def is_file(path):
    return os.path.isfile(path)

def make_dir(path):
    if not os.path.exists(path):
        os.makedirs(path, exist_ok=True)

def move_file(src, dest):
    if DRY_RUN:
        print(f"[DRY-RUN] Would move: {src} -> {dest}")
    else:
        shutil.move(src, dest)
        print(f"Moved: {src} -> {dest}")

def organize_by_date(folder):
    for name in os.listdir(folder):
        full = os.path.join(folder, name)
        if is_file(full):
            # get modification time and format as YYYY-MM-DD
            mtime = os.path.getmtime(full)
            date_str = datetime.fromtimestamp(mtime).strftime("%Y-%m-%d")

            target_folder = os.path.join(folder, date_str)
            make_dir(target_folder)

            dest = os.path.join(target_folder, name)
            # avoid overwrite: if file exists, add counter
            base, ext = os.path.splitext(name)
            counter = 1
            while os.path.exists(dest):
                dest = os.path.join(target_folder, f"{base}_{counter}{ext}")
                counter += 1

            move_file(full, dest)

if __name__ == "__main__":
    if not os.path.isdir(FOLDER_PATH):
        print("ERROR: Folder path does not exist. Edit FOLDER_PATH at top of script.")
    else:
        organize_by_date(FOLDER_PATH)
        print("Done. (DRY_RUN is", DRY_RUN, ")")
```
